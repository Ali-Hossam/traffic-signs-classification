## **Import Libraries**


```python
import numpy as np
import pandas as pd
import PIL
import matplotlib.pyplot as plt
from keras.models import load_model
```

## **Load our data and model**

```python
#Load our data and model
model = load_model("CNN_model.model")
y_train = np.load("y_train.npy")
X_train = np.load("X_train.npy")
```

```python
image_height, image_width = 28, 28
```

## **Classify an image**

```python
def predict_image_class(img_path):
    original_image = PIL.Image.open(image_path)  #load the image to be classified
    im = original_image.resize((image_height, image_width))
    im_data = np.asarray(im)
    im_data = np.expand_dims(im_data, axis=0)
    im_data = im_data / 255

    pred = model.predict(im_data)
    pred = int(pred.argmax(axis=-1))

    #Visualize your image and a random one from the same class
    fig, ax = plt.subplots(1, 2, figsize=(8, 4))
    ax[0].imshow(np.asarray(original_image).astype("uint8"))
    ax[0].set_title(f"The class of your image is {pred}", fontweight="bold")
    ax[0].axis("off")

    y_train_df = pd.DataFrame(y_train)
    random_image_idx = np.random.choice(y_train_df[y_train_df[pred] == 1].index)
    ax[1].imshow(np.asarray(X_train[random_image_idx]))
    ax[1].set_title(f"A random Image from class {pred}", fontweight="bold")
    ax[1].axis("off")

    plt.subplots_adjust(wspace=0.30)
    plt.show()
```

#### To Predict the class of an image put the image in **Images_to_predict** folders, and then put the path of the image in the function **predict_image_class**

```python
#Predict the class of a random image from the web saved locally
image_path = "Imgaes_to_predict/test_image.jpeg"
predict_image_class(image_path)
```
